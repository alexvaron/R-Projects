```python
def isPrime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n%i == 0:
            return False
    return True

import random 

val = random.randint(0,500)
totalSum = 0

for i in range(0, val+1):
    if isPrime(i):
        print(i, " is a prime number")
        totalSum += i
print(totalSum , " is the sum of all prime numbers from 0 to ", val)
```

```python
def maxAdjacentProd(nums):
    maxSoFar = 0
    curProd = 0
    for i in range(0,len(nums)-1):
        curProd = nums[i] * nums[i+1]
        maxSoFar = max(maxSoFar, curProd)
    return maxSoFar

values = [1,1,0,4,10,1,15]
print(maxAdjacentProd(values))
```

```python
import re #regex extension

# Function to categorize
def password(pw):
    # for a newline or space
    if (pw == "\n" or pw == " "):
        return "Password cannot contain a newline or space!"

    # for length in between 9 and 20
    if (9 <= len(pw) <= 20):

        # for occurrence 3+ in a row
        if re.search(r'(.)\1\1', pw):
            return "Weak password! (same character repeats 3+ times in a row)"

        # for occurrence of same string repeating
        if re.search(r'(..)(.*?)\1', pw):
            return "Weak password! (Same string repetition)"
        else:
            return "Strong Password!"
    else:
        return "Password length must be 9-20 characters!"

# Driver code
print(password("AlexandBen"))
print(password("BenBenBen"))
print(password("thispassismorethan20chars"))
print(password("AAABBBOOO"))
```

```python
def lenOfWord(length, words):
    ret = set()
    for word in words:
        if len(word.lower()) == length:
            ret.add(word.lower())
    return ret

words = []
with open('Data Science.txt', 'r') as f:
    for line in f:
        for word in line.split():
            words.append(word)
            
print(lenOfWord(4,words))
print(lenOfWord(3,words))
print(lenOfWord(5,words))
```
